```python
#ETL_sample02.py ETL7のファイルを読み込
# 使用するライブラリを読み込む
import struct
from PIL import Image, ImageEnhance
 
filename = "ETL7/ETL7LC_1"
skip = 100
with open(filename, 'rb') as f:
    f.seek(skip * 2052)
    s = f.read(2052)
    r = struct.unpack('>H2sH6BI4H4B4x2016s4x', s)
    iF = Image.frombytes('F', (64, 63), r[18], 'bit', 4)
#    iP = iF.convert('P')
    iP = iF.convert('L')
    fn = "{:1d}{:4d}{:2x}.png".format(r[0], r[2], r[3])
    iP.save(fn, 'PNG', bits=4)
    enhancer = ImageEnhance.Brightness(iP)
    iE = enhancer.enhance(16)
    iE.save(fn, 'PNG')
```

```python
 #ETL7toimg.py ETL7の4個の分割ファイルを読み，ひらがな画像集を作成
# 使用するライブラリを読み込む
import struct
from PIL import Image, ImageEnhance
import glob, os

# ひらがな画像集を保存するディレクトリ
outdir = "ETL7-img/"
if not os.path.exists(outdir): os.mkdir(outdir)

# ETL7ディレクトリ内部の4個の分割データを読み込む
files = glob.glob("ETL7/*")
for fname in files:
    if fname == "ETL7\ETL7INFO": continue # 不要な分割ファイルは省く
    print(fname) # ETL7の分割ファイル名

    # ETL7の分割ファイル名を開く
    f = open(fname, 'rb')
    f.seek(0)
    while True:
        # あいうえおのラベルと画像データの組を一つずつ読む
        s = f.read(2052)
        if not s: break
        # バイナリデータなのでPythonが理解できるように
        r = struct.unpack('>H2sH6BI4H4B4x2016s4x', s)
        code_ascii = r[1]
        code_jis = r[3]
        # ひらがなの画像として取り出す
        iF = Image.frombytes('F', (64, 63), r[18], 'bit', 4)
        iP = iF.convert('L')
        dir = outdir + "/" + str(code_jis)
        if not os.path.exists(dir): os.mkdir(dir)
        fn = "{0:02x}-{1:02x}{2:04x}.png".format(code_jis, r[0], r[2])
        fullpath = dir + "/" + fn
        #if os.path.exists(fullpath): continue
        enhancer = ImageEnhance.Brightness(iP)
        iE = enhancer.enhance(16)
        iE.save(fullpath, 'PNG')
print("ok")
```

```python
 #ETL7img2data.py 
#画像を読み込み，ひらがな画像データセットを作成
# 使用するライブラリを読み込む
import numpy as np
import cv2
import matplotlib.pyplot as plt
import glob
import pickle

# 保存ディレクトリと画像サイズの指定
out_dir = "./ETL7-img" # ひらがな画像集のディレクトリ
im_size = 32 # 画像サイズ

save_file = out_dir + "/hiragana.pickle" # 保存ファイル名と保存先
plt.figure(figsize=(9, 17)) # 出力画像を大きくする

# ひらがな画像集のディレクトリから画像を読み込み開始
hiraganadir = list(range(177, 220+1)) #あいうえお--わの順序
hiraganadir.append(166) # を
hiraganadir.append(221) # ん
result = []

for i, code in enumerate(hiraganadir):
    img_dir = out_dir + "/" + str(code)
    fs = glob.glob(img_dir + "/*")
    print("dir=",  img_dir)

    # 画像64X63を読み込んでグレイスケールに変換し32X32に整形
    for j, f in enumerate(fs):
        img = cv2.imread(f)
        img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        img = cv2.resize(img_gray, (im_size, im_size))
        result.append([i, img])

        # ひらがな画像一覧表示 10行X5列
        if j == 3:
            plt.subplot(11, 5, i + 1)
            plt.axis("off")
            plt.title(str(i))
            plt.imshow(img, cmap='gray')

# ひらがなの「画像とラベル」のデータセットを保存
pickle.dump(result, open(save_file, "wb"))
plt.show()
```

```python
 #hiragana_cnn-VGG-like.py CNN-VGG-like
# 使用するライブラリを読み込む
import numpy as np
import cv2, pickle
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten
from keras.layers import Conv2D, MaxPooling2D
from keras.optimizers import RMSprop 
from keras.datasets import mnist

#「katakana.pickle」の入力データと出力データを指定
image_rows = 32 # 画像縦の画素数
image_cols = 32 # 画像横の画素数
image_color = 1 # グレースケール
image_shape = (image_rows, image_cols, image_color)  #  画像のデータ形式
image_size = image_rows * image_cols * image_color   #  画像のニューロン数

input_size = image_size # 変換する画像のデータ形式(入力層のニューロン数3072)
output_size = 46 # 画像46分類が目標(出力層のニューロン数46)アイウエオ-ンまでの文字数

data_file = "./ETL7-img/hiragana.pickle" # データファイル

# ひらがな画像のデータセットを読み込む
data = pickle.load(open(data_file, "rb"))
y = [] #ラベルデータ
x = [] #画像データ
for d in data:
    (num, img) = d
    img = img.astype('float').reshape( image_rows, image_cols, image_color) / 255 # 画像データを正規化
    y.append(keras.utils.np_utils.to_categorical(num, output_size)) # ラベルデータをone-hotベクトルに変換
    x.append(img)
x = np.array(x)
y = np.array(y)

# ①②「hiragana.pickle」のデータをロード/データを訓練用とテスト用に分ける 
train_X, test_X,train_Y, test_Y = train_test_split( x, y, test_size = 0.2, train_size = 0.8, shuffle = True)

# CNNモデル構造を定義
# ③分類器「畳み込みニューラルネットワーク(CNN-VGG-like)」を設計

# ③-1 分類器の定義
model = Sequential() # 分類器のインスタンスを生成し，目的に応じたレイヤ(層)を追加する

# ③-1-a 畳み込み層とMaxPooling層を利用した画像の特徴分析

model.add(Conv2D(16, kernel_size=(3, 3), activation='relu', input_shape=image_shape)) # 畳み込みで入力層

model.add(Conv2D(32,(3, 3), activation='relu')) # 畳み込みで入力層

# (入力32X32)->畳み込み層->(出力32)
model.add(Conv2D(64,(3, 3), activation='relu')) # 畳み込みで入力層

# (入力32)->畳み込み層->(出力64)
# model.add(Conv2D(128, (3, 3), activation='relu')) # 畳み込み層を追加

# (入力64)->MaxPooling層->(出力64)
model.add(MaxPooling2D(pool_size=(2, 2))) # MaxPooling層を追加

model.add(Conv2D(128, (3, 3), activation='relu')) # 畳み込み層を追加

# (入力64)->ドロップアウト層->(出力64)  # ドロップアウト層を追加　過学習の防止
model.add(Dropout(0.25))

# ③-1-b 全結合層を使用した画像分類
# (入力64)-> Flatten層->(出力12544)
model.add(Flatten()) # Flatten層　入力を平滑化

# (入力12544)-> 全結合層->(出力128)
model.add(Dense(128, activation='relu'))   # 全結合層を追加

# (入力128)->ドロップアウト層->(出力128)
model.add(Dropout(0.5))   # ドロップアウト層を追加　過学習の防止

# (入力128)->全結合層Dense->(出力46)
model.add(Dense(output_size, activation='softmax')) # 出力層を追加

# ③-2 分類器「畳み込みニューラルネットワーク(CNN-VGG-like)」をコンパイル
model.compile( loss='categorical_crossentropy', optimizer=RMSprop(), metrics=['accuracy'])

# ④分類器「畳み込みニューラルネットワーク(CNN-VGG-like)」でミニバッチを繰り返し訓練を実行
hist = model.fit(train_X, train_Y, batch_size=128, epochs=16, verbose=1, validation_data=(test_X, test_Y))
#hist = model.fit(x_train, y_train, batch_size=128, epochs=12, verbose=1, validation_data=(x_test, y_test))

# ⑤分類器「畳み込みニューラルネットワーク(CNN-VGG-like)」でテストを実行
score = model.evaluate(test_X, test_Y, verbose=1)
print('正解率=', score[1], 'loss=', score[0])

# ⑥結果の出力
# ⑥-1時間をかけた訓練結果のモデル保存
# 場所はカレントディレクトリに保存
model.save('ETL7-CNN-VGG-like_model.h5') # モデル構造と重みパラメータを含む

# ⑥-2 学習経過をグラフに記録
# 正解率の推移をプロット
plt.plot(hist.history['accuracy'])
plt.plot(hist.history['val_accuracy'])
plt.title('Accuracy')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
# ロスの推移をプロット
plt.plot(hist.history['loss'])
plt.plot(hist.history['val_loss'])
plt.title('Loss')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

```python
def show_prediction():
    n_show = 96
    y = model.predict(test_X) # (A)
    plt.figure(2, figsize=(12, 8))
    plt.gray()
    for i in range(n_show):
        plt.subplot(8, 12, i + 1)
        x = test_X[i, :]
        x = x.reshape(32, 32)
        plt.pcolor(1 - x)
        wk = y[i, :]
        prediction = np.argmax(wk)
        plt.text(22, 25.5, "%d" % prediction, fontsize=12)
        if prediction != np.argmax(test_Y[i, :]):
            plt.plot([0, 27], [1, 1], color='cornflowerblue', linewidth=5)
        plt.xlim(0, 27)
        plt.ylim(27, 0)
        plt.xticks([], "")
        plt.yticks([], "")
# -- メイン
show_prediction()
plt.show()
```
